```python
### Design The Sequences ###

from Bio.Seq import Seq
from Bio.SeqUtils import MeltingTemp as mt
import random

#For 7mers
sequences = []
generated_sequences = set()
for i in range(1000):
       sequence = ''
       prev_base = ""
       base_repeated = 0
       #base = ''
       for j in range(7): 
              if j == 0: #can give random bases and specify what bases to be picked random (initialize the base)
                     base = random.choice(['A','G','C','T'])
              elif j != 0:
                     base = random.choice(['A','T','C','G'])
                     while base == prev_base and base_repeated == 1:
                            base = random.choice(['A','T','C','G']) 
              
              if base == prev_base:
                     base_repeated += 1
              else:
                     base_repeated = 0
              
              prev_base = base
              sequence += base
              
       if sequence not in generated_sequences: #ensure no repeated sequences
              sequences.append(sequence)
              generated_sequences.add(sequence)
```

```python
### Compute Tm using Biopython ###

#      '''DEFAULT PARAMETERS''' seq, check=True, strict=True, c_seq=None, shift=0, nn_table=None, tmm_table=None, 
# imm_table=None, de_table=None, dnac1=25, dnac2=25, selfcomp=False, Na=50, K=0, Tris=0, Mg=0, dNTPs=0, saltcorr=5

Tm_16C = []

for sequence in sequences:
       Tm = mt.Tm_NN(sequence, Mg = 10, Na = 5) 
       if 15.80 <= Tm <= 16.70:
              formatted_Tm = "{:.2f}".format(Tm)              
              concat_seq = str(sequence) + ' \t' + str(formatted_Tm)
              Tm_16C.append(concat_seq)

for seq in Tm_16C:
       print(seq)
```
